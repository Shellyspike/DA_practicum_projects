## Предобработка данных

### Обработка пропусков

Представляется, что легче всего заполнить пропуски в следующих столбцах: является ли квартира апартаментами, наличие балкона (и возможно кухни), информация о аэропортах, парках и прудах рядом с квартирой.

Начнем со столбца is_apartment. Проверим значения в столбце.

```python
df['is_apartment'].value_counts()
```

Так как апартаменты - это относящиеся к нежилому фонду жилые помещения, можно предположить, что встречаются они достаточно редко, а заполнение данного пункта является необязательным для пользователя, поэтому его часто пропускают. Заполним пропуски в данном столбце значениями False и проверим результат.

```python
df['is_apartment'] = df['is_apartment'].fillna(value=False)
df['is_apartment'].value_counts()
```

Изучим пропуски в столбцах с информацией о балконах и кухне.

```python
print(f'Данные о наличии балкона отсутствуют в {df["balcony"].isna().sum()} случаях')
display()
print(f'Данные о площади кухни отсутствуют в {df["kitchen_area"].isna().sum()} случаях')
```

Квартиры без кухни встречаются значительно реже, чем квартиры без балкона. Можно предположить, что в обоих случах данные пропущены в связи с отсутствием в квартире балкона или кухни и можно заполнить данные значениями 0.

Проверим, сколько из квартир без кухни являются также студиями или квартирами с открытой планировкой.

```python
df.loc[df['kitchen_area'].isna()&df['studio']==True][['kitchen_area', 'studio', 'total_area', 'living_area']]
```

```python
len(df.loc[df['studio']==True])
```

Во всех случаях, когда квартира указана как студия, нет кухни.

```python
df.loc[df['kitchen_area'].isna()&df['open_plan']==True][['kitchen_area', 'open_plan', 'total_area']]
```

```python
len(df.loc[df['open_plan']==True])
```

Также во всех случаях, когда указано, что квартира с открытой планировкой, кухня отсутствует.

Студий и квартир с открытой планировкой значительно меньше, чем пропусков в столбце с данными о площади кухни. Для большей точности сделаем следующее:
- заполним данные о балконах и площади кухни, там где она точно отсутствует, нулями;
- в остальных случаях посчитаем среднее процентное соотношение кухни и общей площади и заполним пропуски этим значением.

```python
(df['kitchen_area']*100/df['total_area']).median().round(2)
```

```python
df['kitchen_area'] = df.apply(
lambda row: 0 if np.isnan(row['kitchen_area']) and any([row['open_plan'], row['studio']]) is True else row['kitchen_area'],
axis=1
)
```

```python
df['kitchen_area'] = df.apply(
    lambda row: row['total_area']*0.17 if np.isnan(row['kitchen_area']) else row['kitchen_area'],
    axis=1
)
```

```python
df['balcony'] = df['balcony'].fillna(value=0)

print(f'Данные о наличии балкона отсутствуют в {df["balcony"].isna().sum()} случаях')
display()
print(f'Данные о площади кухни отсутствуют в {df["kitchen_area"].isna().sum()} случаях')
```

Проверим, не оказалась ли где-то площадь кухни больше возможной.

```python
len(df.loc[df['total_area']-df['living_area']-df['kitchen_area']<0])
```

Здесь тоже поставим 0 вместо площади кухни.

```python
df['kitchen_area'] = df.apply(
    lambda row: 0 if row['total_area']-row['living_area']-row['kitchen_area']<0 else row['kitchen_area'],
    axis=1
)
```

Можно предположить, что данные о расстоянии до парка и водоемов пропущены в тех случаях, когда ни парков, ни водоемов рядом нет. В этом случае пропуски заполнять не следует. Проверим, так ли это.

```python
df.loc[df['ponds_around3000'].isna()&df['ponds_around3000'].isna()].head()
```

Из обеих таблиц видно, что если обозначить квартиру на карте, автоматически заполняется информация о расстоянии до важных объектов: аэропортов, парков, водоемов и центра города. 

Если в этом случае заполнить или отбросить все пропущенные значения, информация в таблице может исказиться. 

Чтобы данные о количестве парков и водоемов поблизости выглядели нормально, надо далее поменять их тип на целочисленный. Для этого здесь заполним пропуски нулями. Т.к. они не были указаны в объявлении, на формирование цены они не влияли - можно считать, что цена формировалась так, будто их не было.

```python
df['parks_around3000'] = df['parks_around3000'].fillna(value=0)
df['ponds_around3000'] = df['ponds_around3000'].fillna(value=0)
```

Также далее надо поменять на целочисленные значения в столбце с количеством этажей. Пропусков там мало, однако они могут повлиять на статистику, если учесть, что цена квартиры на самом верхнем этаже обычно ниже. 

Подставим в столбец значения из заполненного полностью столбца с информацией о этаже, на котором находится квартира, но вычтем из него 1, чтобы при классификации по этажам можно было отнести эти квартиры к категории "Другой этаж".

```python
df['floors_total'] = df['floors_total'].fillna(value=df['floor']-1)
```

Также можно заполнить пропуски в столбце с жилой площадью, рассортировав квартиры по общей площади. Сначала проверим, нет ли в используемых столбцах подозрительных значений.

```python
df.total_area.hist()
```

```python
df.living_area.hist()
```

```python
df.total_area.hist(range=(0,200))
df.living_area.hist(range=(0,200), alpha=0.5)
plt.legend(['total_area','living_area'])
plt.show()
```

В действительности площади квартир распределяются примерно также: много квартир средней площади, чем больше площадь, тем меньше количество таких квартир. Строк, где жилая площадь больше общей, также нет.

Заполним пропущенные значения, классифицировав квартиры по площади.

```python
cut_labels = ['small', 'medium', 'big', 'elite']
cut_bins = [0, 45, 75, 110, 900]

df['total_area_groups'] = pd.cut(df['total_area'],
                       bins=cut_bins,
                       labels=cut_labels)
```

```python
df['living_area'] = df['living_area'].fillna(df.groupby('total_area_groups')['living_area'].transform('median'))
```

Проверим, не получилось ли так, что где-то жилая площадь стала меньше 0 и удалось ли заполнить пропуски.

```python
display(len(df.loc[df['living_area']<0]))
display()
display(df['living_area'].isna().sum())
```

```python
df.isna().sum()
```

```python
df.isna().mean()
```

Оставшиеся пропуски невозможно заполнить по следующим причине того, что их нельзя как-то классифицировать в целях заполнения. Высота потолков может не зависеть от общей площади. Также с общим количеством этажей - в одном районе города могут находится здания различной высоты.

Также с географическим положением квартир. Зная расстояние до парков, водоемов и аэропортов возможно построить модель определения местоположения квартиры, но в данном случае для большого города и с достаточно большим количеством пропусков она будет неточной.
