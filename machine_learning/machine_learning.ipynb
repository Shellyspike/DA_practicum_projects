## Информация об исследовании и обрабатываемых данных

Целью данного проекта является анализ анкет клиентов фитнес-центра для прогнозирования вероятности оттока клиентов. Для этого необходимо выделить основные группы клиентов и сформулировать признаки, влияющие на их отток.

Данные представлены следующим датасетом с информацией о клиентах:

**Данные клиента за предыдущий до проверки факта оттока месяц:**
* gender — пол;
* Near_Location — проживание или работа в районе, где находится фитнес-центр;
* Partner — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент);
* Promo_friends — факт первоначальной записи в рамках акции «приведи друга»;
* Phone — наличие контактного телефона;
* Age — возраст;
* Lifetime — время с момента первого обращения в фитнес-центр (в месяцах).

**Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:**

* Contract_period — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
* Month_to_end_contract — срок до окончания текущего действующего абонемента (в месяцах);
* Group_visits — факт посещения групповых занятий;
* Avg_class_frequency_total — средняя частота посещений в неделю за все время с начала действия абонемента;
* Avg_class_frequency_current_month — средняя частота посещений в неделю за предыдущий месяц;
* Avg_additional_charges_total — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.
* Churn — факт оттока в текущем месяце.

В результате изучения данных будут даны рекомендации по повышению качества работы с клиентами и предложены меры по снижению оттока клиентов.

## Построение модели прогнозирования оттока клиентов

### Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце.

Сначала удалим по одному из сильно коррелирующих друг с другом признаков. Оставим количество посещений в целом и период, на который заключен контракт.

```python
clients_corrected = clients.copy()
clients_corrected = clients_corrected.drop(['month_to_end_contract', 'avg_class_frequency_current_month'], axis=1)
```

Разделим таблицу на признаки и целевую переменную - churn.

```python
X = clients_corrected.drop('churn', axis = 1)
y = clients_corrected['churn']
```

Разобьем данные на обучающие и валидационные выборки.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, shuffle=False)
```

Поскольку диапазоны значений признаков, представленных в таблице, значительно отличаются, их необходимо стандартизировать.

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.transform(X_train) 
X_test_st =  scaler.transform(X_test)
```

Проверим правильности разбиения.

```python
for df in [X_train, X_test, y_train, y_test]:
    display(df.shape)
```

800 и 3200 строк - это 20% и 80% от 4000. Разбиение произведено правильно.

### Обучим полученную модель и сравним метрики на валидационной выборке

Для каждой модели будут сравниваться следующие метрики:
* accuracy_score - доля правильных ответов,
* precision_score - доля правильных ответов среди целевого класса,
* recall_score - доля реальных объектов целевого класса, обнаруженных с помощью модели,
* f1_score - баланс между двумя предыдущими метриками
* roc_auc_score - оценка качества классификатора

```python
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

1. Логистическая регрессия без стандартизации

```python
lr_model = LogisticRegression(solver='liblinear')
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
lr_probabilities = lr_model.predict_proba(X_test)[:, 1]
```

2. Случайный лес без стандартизации

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_probabilities = rf_model.predict_proba(X_test)[:, 1]
```

```python
print_all_metrics(y_test, lr_predictions, lr_probabilities, title='Метрики для модели логистической регрессии без стандартизации:')
print_all_metrics(y_test, rf_predictions, rf_probabilities, title='Метрики для модели случайного леса без стандартизации:')
```

Сделаем расчеты по стандартизированным данным

1. Логистическая регрессия

```python
lr_model = LogisticRegression(solver='liblinear')
lr_model.fit(X_train_st, y_train)
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
```

2. Случайный лес

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
rf_model.fit(X_train_st, y_train)
rf_predictions = rf_model.predict(X_test_st) 
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
```

```python
print_all_metrics(y_test, lr_predictions, lr_probabilities, title='Метрики для модели логистической регрессии со стандартизацией:')
print_all_metrics(y_test, rf_predictions, rf_probabilities, title='Метрики для модели случайного леса со стандартизацией:')
```

Со стандартизацией модели показывают себя чуть лучше. Также результат становится лучше, если не перемешивать данные перед разбивкой на выборки. 

Модель логистической регрессии со стандартизацией показывает себя лучше всего, а также является более точной по результатам метрики roc_auc_score.

## Кластеризация клиентов

Для кластеризации уберем из таблицы столбец с оттоком клиентов.

```python
clients_cl = clients_corrected.drop('churn', axis = 1)
```

Проведем стандартизацию данных.

```python
clients_cl = scaler.fit_transform(clients_cl)
```

### Построим матрицу расстояний на стандартизованной матрице признаков и нарисуем дендрограмму, чтобы определить, какое количество кластеров можно выделить.

```python
linked = linkage(clients_cl, method = 'ward')
```

```python
plt.figure(figsize=(23, 9))  
dendrogram(linked, orientation='top', no_labels=True)
plt.title('Кластеризация клиентов', size=20)
plt.show() 
```

Судя по соотношению размера групп, можно выделить 5 кластеров, т.к. 4 группа получилась достаточно большой. 

### Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. 

На основании построенной выше дендрограммы укажем 5 в качестве числа кластеров.

```python
km = KMeans(n_clusters=5, random_state=0)
labels = km.fit_predict(clients_cl)
```

Сохраним метки кластера в поле датасета.

```python
clients['cluster_km'] = labels
```

Изучим средние значения признаков для кластеров. Для этого сгруппируем датасет по кластеру.

```python
clients.groupby('cluster_km').mean().T
```

Больше всего отток в кластере 0 и кластере 2. В кластере 0 нет никого, кто жил бы рядом с фитнес-центром. В классе 2 меньше всего клиентов, пришедших по промо от друзей и от партнеров клуба. В обоих случаях срок контракта меньше, чем у других кластеров, а также меньше количество посещений.

### Построим распределения признаков для кластеров.

```python
for column in ['gender', 'near_location', 'partner', 'promo_friends', 'phone', 'group_visits']:
    plt.figure(figsize=(6, 6))
    sns.countplot(x = clients[column], hue='cluster_km', data=clients, palette='pastel')
    plt.title(column)
    plt.show()
```

```python
clients['cluster_km'] = clients['cluster_km'].apply(str)
for column in ['age', 'contract_period', 'month_to_end_contract', 'lifetime', 'avg_class_frequency_total', 'avg_class_frequency_current_month', 'avg_additional_charges_total']:
    plt.figure(figsize=(6, 6))
    plt.title(column)   
    sns.kdeplot(x = clients[column], hue='cluster_km', data=clients, palette='pastel')
    plt.show()
```

По данным графиков можно выделить следующие особенности клиентов кластера 0 и кластера 2, где наблюдается самый большой отток:
* 0: живут далеко, нет промо от друга, не посещают групповые занятия, заключают контракт на короткий срок, обратились недавно, посещают в среднем 1,5 занятия в неделю, тратят меньше других
* 2: живут рядом, нет промо от друга или партнера, не посещают групповые занятия, заключают контракт на короткий срок, посещают 2 занятия в неделю

### Рассчитаем долю оттока клиентов для каждого кластера

```python
clients.groupby('cluster_km').agg({'churn':'mean'}).sort_values(by='churn')
```

Как уже было отмечено ранее, самый большой отток в кластере 0, за ним следует кластер 2. 

Менее всего склонен к оттоку кластер 1. Судя по графикам, в него входят посетители, которые живут рядом, пришли в фитнес-центр по промо-акции или от партнера. Также здесь больше, чем в других кластерах, людей, которые заключают контракт на длительный срок

По результатам исследования можно сделать вывод, что оттоку подвержено достаточно большое количество клиентов фитнес-центра - 25%. Необходимо усовершенствовать политику работы с клиентами, чтобы снизить долю оттока. 

По данным кластера, менее всего склонного к оттоку и 2 кластеров, откуда уходит больше всего людей, можно дать следующие рекомендации:
1. сделать бонусы для тех, кто заключает контракт на длительный срок
2. мотивировать клиентов к более частому посещению занятий - например добавлять дополнительные дни к контракту за определенное количество посещений в месяц
3. изучить наличие недостатков у групповых занятий - большинство клиентов из нестабильных кластеров их не посещает
4. мотивировать клиентов, заключивших пробный контракт на короткий срок, продолжать занятия - например предоставлять им бесплатную персональную тренировку и консультацию врача, где им дадут рекомендации по определению программы занятий
